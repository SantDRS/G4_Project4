```python
# Import csv's & load df's
# Denver
denver_file_path = '/Users/rickygarcia/ModuleChallenges/G4_Project4/Resources/df_Extractions/Denver_df.csv'

denver_df = pd.read_csv(denver_file_path)

denver_df.head(2)
```

```python
# Import csv's & load df's
# Miami
miami_file_path = '/Users/rickygarcia/ModuleChallenges/G4_Project4/Resources/df_Extractions/Miami_df.csv'

miami_df = pd.read_csv(miami_file_path)

miami_df.head(2)
```

```python
# Import csv's & load df's
# NewYork
newyork_file_path = '/Users/rickygarcia/ModuleChallenges/G4_Project4/Resources/df_Extractions/NewYork_df.csv'

newyork_df = pd.read_csv(newyork_file_path)

newyork_df.head(2)
```

```python
print(denver_df.info())
print(miami_df.info())
print(newyork_df.info())
```

```python
import pandas as pd

# Counting number of categories in denver_df
denver_unique_categories = denver_df['categories'].explode().unique()
den_num_unique_categories = len(denver_unique_categories)
print("Denver number of unique categories:", den_num_unique_categories)

# Counting number of categories in miami_df
miami_unique_categories = miami_df['categories'].explode().unique()
mia_num_unique_categories = len(miami_unique_categories)
print("Miami number of unique categories:", mia_num_unique_categories)

# Counting number of categories in newyork_df
ny_unique_categories = newyork_df['categories'].explode().unique()
ny_num_unique_categories = len(ny_unique_categories)
print("New York number of unique categories:", ny_num_unique_categories)
```

```python
# Merging all Dataframes

merge_df = pd.concat([denver_df, miami_df, newyork_df], ignore_index = True)
merge_df
```

```python
# Counting number of categories in merge_df

merge_df_categories = merge_df['categories'].explode().unique()
merge_df_unique_categories = len(merge_df_categories)
print("The merged data frame number of unique categories:", merge_df_unique_categories)
```

```python
# Extract df

output_path = '/Users/rickygarcia/ModuleChallenges/G4_Project4/Resources/df_Extractions'
output_file = os.path.join(output_path, 'merged_df.csv')

# Export DataFrame to CSV
merge_df.to_csv(output_file, index=False)

print("DataFrame extracted and saved as CSV successfully.")
```
