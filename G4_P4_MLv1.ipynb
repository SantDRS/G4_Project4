```python
import json
import os
import pandas as pd
import pprint as pp
import requests
import time
#from tensorflow.keras.models import load_model
#import warnings
#warnings.simplefilter("ignore")
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Import csv's & load df's
# Denver
merged_file_path = '/Users/josesantos/LearnPython/BCGit/G4_Project4/Resources/df_Extractions/merged_df.csv'

merged_df = pd.read_csv(merged_file_path)

merged_df.head(2)
```

```python
columns_to_delete = ['location', 'coordinates']
merged_df.drop(columns_to_delete, axis=1, inplace=True)

# Access and work with the updated DataFrame
merged_df.head(1)
```

```python
merged_df = merged_df[['name', 'categories', 'rating', 
                       'review_count', 'city']]

# Access and work with the selected DataFrame
merged_df
```

```python
# City - One-hot encoding

city_encoded_data = pd.get_dummies(merged_df, columns = ['city'])
city_encoded_data
```

```python
city_encoded_data.info()
```

```python
# Ratings - labelEncoder
ratings = city_encoded_data['rating']

encoder = LabelEncoder()
encoder.fit(ratings)

encoded_ratings = encoder.transform(ratings)
city_encoded_data['rating_encoded'] = encoded_ratings

city_encoded_data
```

```python
# Review_count - MinMaxScaler

scaler = MinMaxScaler()
review_counts = city_encoded_data['review_count'].values.reshape(-1,1)
normalized_review_counts = scaler.fit_transform(review_counts)
city_encoded_data['normalized_review_count'] = normalized_review_counts

city_encoded_data
```

```python
bins = [-1, 0.33, 0.66, 1.0]
labels = ['low', 'medium', 'high']

city_encoded_data['review_count_category'] = pd.cut(city_encoded_data['normalized_review_count'], bins=bins, labels=labels)

city_encoded_data
```

```python
# Categories - labelEncoder
# Explode the 'categories' column
city_encoded_data['categories'] = city_encoded_data['categories'].apply(eval)

city_encoded_data = city_encoded_data.explode('categories')

city_encoded_data
```

```python
# Create a label encoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'categories' column
city_encoded_data['categories_encoded'] = label_encoder.fit_transform(city_encoded_data['categories'])

city_encoded_data
```

```python
city_encoded_df = city_encoded_data[['name', 'city_Denver', 'city_Miami', 'city_NewYork', 'rating_encoded', 'normalized_review_count', 'categories_encoded']]
city_encoded_df
```
